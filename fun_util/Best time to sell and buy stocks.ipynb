<h1>Best time to sell and buy stocks</h1>

```python
from typing import List

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        maxprofit = 0
        for i in range(1, len(prices)):
            if (prices[i] > prices[i-1]):
                maxprofit += prices[i] - prices[i-1]
            
        return maxprofit
```

```python
s = Solution()
prices = [7,1,5,3,6,4]
p = s.maxProfit(prices)
print(p)
```
