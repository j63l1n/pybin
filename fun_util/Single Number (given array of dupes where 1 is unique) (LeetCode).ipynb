<h1>Given a non-empty array of integers (nums), every element appears twice except for one. Find that single one</h1>

```
Example 1:
Input: nums = [2,2,1]
Output: 1
```

using xor, where a gets to be either 0 or 1  so we can sum them together to get the unique one...

````python
class Solution(object):
    def singleNumber(self, nums):
        a = 0
        for i in nums:
            a ^= i
        return a
````

````python
s = Solution()
nums = [1,2,4,1,2]
s.singleNumber(nums)
````
